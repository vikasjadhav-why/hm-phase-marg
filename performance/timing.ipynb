```python
quad = h5py.File(f'../outputs/mode_2233/quad/noise_hm/out_files/time.hdf','r')['times']
grid = h5py.File(f'../outputs/mode_2233/grid/noise_hm/out_files/time.hdf','r')['times']
analytic_all = h5py.File(f'../outputs/mode_2233/peaks_all/noise_hm/out_files/time.hdf','r')['times']
analytic_22 = h5py.File(f'../outputs/mode_2233/peaks_dominant/noise_hm/out_files/time.hdf','r')['times']
analytic_all_vec = h5py.File(f'../outputs/mode_2233/vec_all/hm/out_files/time.hdf','r')['times']
analytic_22_vec = h5py.File(f'../outputs/mode_2233/vec_dominant/hm/out_files/time.hdf','r')['times']
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set publication-quality defaults
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = ['DejaVu Serif', 'Times New Roman', 'serif']
mpl.rcParams['mathtext.fontset'] = 'dejavuserif'  # For math symbols
mpl.rcParams['font.size'] = 11
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['axes.titlesize'] = 12
mpl.rcParams['xtick.labelsize'] = 10
mpl.rcParams['ytick.labelsize'] = 10
mpl.rcParams['legend.fontsize'] = 10
mpl.rcParams['figure.titlesize'] = 12
mpl.rcParams['lines.linewidth'] = 1.5

# Create figure with single column width
fig, ax = plt.subplots(figsize=(7, 7),dpi=300)

# Plot histograms with distinct colors and transparency
# Using step histograms for better overlap visibility
#ax.hist(quad, bins=30, label='Quadrature',
#        alpha=0.6, color='#E41A1C', edgecolor='#E41A1C', 
#        linewidth=1.5, histtype='stepfilled')

ax.hist(grid, bins=30, label='Grid',
        alpha=0.4, color='#377EB8', edgecolor='#377EB8',
        linewidth=1.5, histtype='stepfilled')

#ax.hist(analytic_22, bins=30, label='Analytic (dominant)',
#        alpha=0.4, color='#4DAF4A', edgecolor='#4DAF4A',
#        linewidth=1.5, histtype='stepfilled')

#ax.hist(analytic_all, bins=30, label='Analytic (all)',
#        alpha=0.4, color='#984EA3', edgecolor='#984EA3',
#        linewidth=1.5, histtype='stepfilled')

ax.hist(analytic_all_vec, bins=30, label='Vectorized Analytic (all)',
        alpha=0.4, color='#FF7F00', edgecolor='#984EA3',
        linewidth=1.5, histtype='stepfilled')

ax.hist(analytic_22_vec, bins=30, label='Vectorized Analytic (dominant)',
        alpha=0.4, color='#A65628', edgecolor='#984EA3',
        linewidth=1.5, histtype='stepfilled')

# Labels
ax.set_xlabel(r'$T (ms)$')
ax.set_ylabel(r'$Counts$')
ax.set_xscale('log')
# Legend with frame
ax.legend(frameon=True, fancybox=False, edgecolor='black', 
          framealpha=1.0, loc='upper left')

# Optional: Add minor ticks for more professional look
ax.minorticks_on()

# Tighten layout
plt.tight_layout()


#plt.savefig('vec_analytic_grid.png', dpi=300, bbox_inches='tight')

plt.show()
```
